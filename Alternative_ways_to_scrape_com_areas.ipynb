## Alternative solutions for a well-known problem

```python
import pandas as pd
import ssl

ssl._create_default_https_context  = ssl._create_unverified_context
multi_tables = pd.read_html('https://en.wikipedia.org/wiki/Community_areas_in_Chicago')
```

```python
print(len(multi_tables))

multi_tables[0]
```

```python
df = multi_tables[0]

Chicago_com_df_v2 = df[["No.", "Name"]]
Chicago_com_df_v2.columns = ["area_code", "community_name"]

Chicago_com_No_nums = Chicago_com_df_v2["area_code"].drop_duplicates().reset_index(drop = True)

Chicago_com_df_v2 = pd.DataFrame(
    {
        "area_code" : range(1, len(Chicago_com_No_nums)),
        "community_name" : Chicago_com_df_v2["community_name"][:-1]
    }
)

Chicago_com_df_v2
```

```python
df = multi_tables[0]

Chicago_com_df_v2 = df[["No.", "Name"]]
Chicago_com_df_v2.columns = ["area_code", "community_name"]

Chicago_com_df_v2["area_code"] = pd.to_numeric(Chicago_com_df_v2["area_code"], 
                                               errors = "coerce")
Chicago_com_df_v2 = Chicago_com_df_v2.dropna(subset = ["area_code"])
Chicago_com_df_v2.reset_index(drop = True, 
                              inplace = True)

Chicago_com_df_v2["area_code"] = Chicago_com_df_v2["area_code"].astype(int)

Chicago_com_df_v2
```

```python
Chicago_com_df_v2.to_csv("Chicago_community_areas_master_v2.csv",
                      index = False)
```
