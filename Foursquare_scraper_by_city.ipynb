```python
from bs4 import BeautifulSoup

import ssl
import urllib.request, urllib.parse
import pandas as pd
```

```python
def város_ismertető():
    ssl._create_default_https_context = ssl._create_unverified_context
    városnév = input("Írd be, melyik város érdekes helyeire lennél kíváncsi: ")
    url = f"https://foursquare.com/explore?mode=url&near={urllib.parse.quote(városnév)}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7"}
    request = urllib.request.Request(url, None, headers)
    response = urllib.request.urlopen(request)
    data = response.read()
    soup = BeautifulSoup(data, "html.parser")

    # A korábbiaknak megfelelően hozom létre az elemek objektumot
    elemek = soup.find_all("div", {"class": "contentHolder"})
    data_list = []
    ertekeles = {}

    
    # Ez az utolsó házival jóformán megegyező ciklus, ami érdemben elkülönül, hogy ebben már i-re és j-re is
    # iterálunk, az ehhez eszközt adó "enumerate"-tel az elemek listán
    for i, j in enumerate(elemek):
        
        hely_neve_j = j.find("div", class_="venueName")
        if hely_neve_j:
            hely_neve = hely_neve_j.text.strip()
        else:
            hely_neve = "A kategóriában nem értelmezhető érték"
        
        kategoria_j = j.find("span", class_="categoryName")
        if kategoria_j:
            kategoria = kategoria_j.text.strip()
        else:
            kategoria = "A kategóriában nem értelmezhető érték"
        
        arazasa_j = j.find("span", itemprop="priceRange")
        if arazasa_j:
            arazasa = arazasa_j.text.strip()
        else:
            arazasa = "A kategóriában nem értelmezhető érték"
        
        cime_j = j.find("div", class_="venueAddress")
        if cime_j:
            cime = cime_j.text.strip()
        else:
            cime = "A kategóriában nem értelmezhető érték"
        
        ertekelese_j = j.find("div", class_="venueScore")
        if ertekelese_j:
            ertekelese = ertekelese_j.text.strip()
        else:
            ertekelese = "A kategóriában nem értelmezhető érték"
        
        data_list.append([i + 1, hely_neve, kategoria, arazasa, cime, ertekelese])
        
        
        #
        try:
            txt = j.find("p")
            if txt:
                review_text = list(txt.children)[1]
            else:
                review_text = "Ehhez a helyhez nincsen szöveges értékelő"
            ertekeles[i + 1] = review_text
        except:
            ertekeles[i + 1] = None


    # 
    df = pd.DataFrame(data_list, columns=["helyezés",
                                          "név",
                                          "típus",
                                          "ár",
                                          "cím",
                                          "score"])
    
    #
    df["szöveges"] = df["helyezés"].map(ertekeles)
    
    # A kérésnek megfelelő adatbázist adja vissza a function
    return df
```

```python
# https://foursquare.com/explore?mode=url&near=Szeksz%C3%A1rd%2C%20Hungary&nearGeoId=72057594040972696
# szekszard

város_ismertető()
```
