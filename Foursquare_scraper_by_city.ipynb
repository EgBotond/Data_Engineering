```python
from bs4 import BeautifulSoup

import ssl
import urllib.request, urllib.parse
import pandas as pd
```

```python
def v√°ros_ismertet≈ë():
    ssl._create_default_https_context = ssl._create_unverified_context
    v√°rosn√©v = input("√çrd be, melyik v√°ros √©rdekes helyeire lenn√©l k√≠v√°ncsi: ")
    url = f"https://foursquare.com/explore?mode=url&near={urllib.parse.quote(v√°rosn√©v)}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7"}
    request = urllib.request.Request(url, None, headers)
    response = urllib.request.urlopen(request)
    data = response.read()
    soup = BeautifulSoup(data, "html.parser")

    # A kor√°bbiaknak megfelel≈ëen hozom l√©tre az elemek objektumot
    elemek = soup.find_all("div", {"class": "contentHolder"})
    data_list = []
    ertekeles = {}

    
    # Ez az utols√≥ h√°zival j√≥form√°n megegyez≈ë ciklus, ami √©rdemben elk√ºl√∂n√ºl, hogy ebben m√°r i-re √©s j-re is
    # iter√°lunk, az ehhez eszk√∂zt ad√≥ "enumerate"-tel az elemek list√°n
    for i, j in enumerate(elemek):
        
        hely_neve_j = j.find("div", class_="venueName")
        if hely_neve_j:
            hely_neve = hely_neve_j.text.strip()
        else:
            hely_neve = "A kateg√≥ri√°ban nem √©rtelmezhet≈ë √©rt√©k"
        
        kategoria_j = j.find("span", class_="categoryName")
        if kategoria_j:
            kategoria = kategoria_j.text.strip()
        else:
            kategoria = "A kateg√≥ri√°ban nem √©rtelmezhet≈ë √©rt√©k"
        
        arazasa_j = j.find("span", itemprop="priceRange")
        if arazasa_j:
            arazasa = arazasa_j.text.strip()
        else:
            arazasa = "A kateg√≥ri√°ban nem √©rtelmezhet≈ë √©rt√©k"
        
        cime_j = j.find("div", class_="venueAddress")
        if cime_j:
            cime = cime_j.text.strip()
        else:
            cime = "A kateg√≥ri√°ban nem √©rtelmezhet≈ë √©rt√©k"
        
        ertekelese_j = j.find("div", class_="venueScore")
        if ertekelese_j:
            ertekelese = ertekelese_j.text.strip()
        else:
            ertekelese = "A kateg√≥ri√°ban nem √©rtelmezhet≈ë √©rt√©k"
        
        data_list.append([i + 1, hely_neve, kategoria, arazasa, cime, ertekelese])
        
        
        #
        try:
            txt = j.find("p")
            if txt:
                review_text = list(txt.children)[1]
            else:
                review_text = "Ehhez a helyhez nincsen sz√∂veges √©rt√©kel≈ë"
            ertekeles[i + 1] = review_text
        except:
            ertekeles[i + 1] = None


    # 
    df = pd.DataFrame(data_list, columns=["helyez√©s",
                                          "n√©v",
                                          "t√≠pus",
                                          "√°r",
                                          "c√≠m",
                                          "score"])
    
    #
    df["sz√∂veges"] = df["helyez√©s"].map(ertekeles)
    
    # A k√©r√©snek megfelel≈ë adatb√°zist adja vissza a function
    return df
```
